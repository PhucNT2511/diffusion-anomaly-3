```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path1 = 'D:/medical_DF/diffusion-anomaly-vanilla/classifier_loss_statistics/training_losses.npy'
path2 = 'D:/medical_DF/diffusion-anomaly-vanilla/classifier_loss_statistics/1000_step_validation.npy'
```

```python
data1 = np.load(path1)
```

```python
data1
```

```python
# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(data1, marker='o', color='b', label='Training Accuracy')
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.title('Training Accuracy Over Iterations')
plt.legend()
plt.grid(True)
plt.show()
```

```python
data2 = np.load(path2,allow_pickle=True).item()
```

```python
val_loss = data2['loss']
val_acc = data2['acc']
```

```python
# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(data1, marker='o', color='b', label='Training Accuracy')
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.title('Training Accuracy Over Iterations')
plt.legend()
plt.grid(True)
plt.show()
```

```python
iterations = np.arange(0, len(val_loss) * 1000, 1000)

# Plotting the data
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot val_loss with the first y-axis
ax1.plot(iterations, val_loss, color='b', label='Validation Loss')
ax1.set_xlabel('Iteration')
ax1.set_ylabel('Validation Loss', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Create a second y-axis for val_acc
ax2 = ax1.twinx()
ax2.plot(iterations, val_acc, color='g', label='Validation Accuracy')
ax2.set_ylabel('Validation Accuracy', color='g')
ax2.tick_params(axis='y', labelcolor='g')

# Add legends
fig.tight_layout()
fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
plt.title('Validation Loss and Accuracy Over Iterations')
plt.grid(True)
plt.show()
```
